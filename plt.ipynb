```python
import json
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python

with open('./repro_lengths.json','r') as f:
        lengths = json.load(f)  
data = defaultdict(int)
for value in lengths.values():
    data[value] += 1

x = list(data.keys())
y = list(data.values())
plt.bar(x, y)
plt.xlabel(f'Reduced Syscall Sequence Length')
plt.ylabel('Count')
plt.title(f'Sequence Length Distribution (total:{len(lengths.values())})')
#plt.yticks(range(int(min(y)), int(max(y)) + 1))
if max(y) - min(y) > 20: 
    y_ticks = np.arange(0, max(y) + 1, step=max(y) // 10)
    plt.yticks(np.append(np.delete(y_ticks,-1),max(y)))
else:
    plt.yticks(range(int(min(y)), int(max(y)) + 1))
    
x_ticks = np.arange(10, max(x)+1, step=(max(x)-10) // 5)
plt.xticks(np.append(range(1,10, 3), x_ticks))
#plt.xticks([i for i in x if i != 0])
plt.savefig(f'C:/Users/Shaw/Desktop/repoducers/fig/sequence_length_distribution_{len(lengths.values())}.png')
plt.show()
```
